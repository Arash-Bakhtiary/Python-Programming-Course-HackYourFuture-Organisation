Question: Build a program to calculate shipping fees for an online store.

- Ask the user for their total order value and the destination country.

- If the order value is over $100, offer free shipping.

- If the order value is between $50  and $100, apply a 5 percentage shipping fee.

- For international orders, apply a flat $20 shipping fee.

- For domestic orders (Canada and US), apply a $10 shipping fee.

```python
order_value = float(input("Enter the total order value: "))
destination = input("Enter the destination country: ")

if order_value > 100:
    shipping_fee = 0
elif 50 <= order_value <= 100:
    shipping_fee = 0.05 * order_value
else:
    shipping_fee = 0

if destination.lower() == "us" or destination.lower() == "canada":
    shipping_fee += 10
else:
    shipping_fee += 20

print(f"Shipping fee: ${shipping_fee:.2f}")
```

- If the user is between 12 and 18 years old, they can watch G, PG, and PG-13 movies.

Question: Create a program that simulates a basic shopping cart with discounts.

- Ask the user for the total cost of items in their cart.

- If the total cost is over $100, offer a 10% discount.

- If the user has a coupon code "SALE20", apply an additional 20% discount.

- Finally, ask the user if they are a premium member. If yes, give them free shipping; otherwise,  calculate shipping costs of $5.

```python
total_cost = float(input("Enter the total cost of items: "))
coupon_code = input("Enter coupon code (if any): ")
premium_member = input("Are you a premium member? (yes/no): ").lower() == "yes"

if total_cost > 100:
    total_cost -= 0.10 * total_cost  # 10% discount

if coupon_code == "SALE20":
    total_cost -= 0.20 * total_cost  # Additional 20% discount

if not premium_member:
    shipping_cost = 5
    total_cost += shipping_cost

print(f"Total cost: ${total_cost:.2f}")
```

Question: Create a program to find prime numbers in a given range using list comprehension.

- Ask the user for a start and end value of the range.
- Generate a list of prime numbers within that range using list comprehension.
- Display the list of prime numbers.

```python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

start = int(input("Enter the start value: "))
end = int(input("Enter the end value: "))

prime_numbers = [num for num in range(start, end + 1) if is_prime(num)]

print("Prime numbers:", prime_numbers)
```

Question: Write a program to categorize words based on their lengths using list comprehension.

- Ask the user for a sentence.
- Generate a dictionary where keys represent word lengths and values are lists of words of that length.
- Display the dictionary.

```python
sentence = input("Enter a sentence: ")
words = sentence.split()

word_lengths = {length: [word for word in words if len(word) == length] for length in range(1, max(map(len, words)) + 1)}

for length, word_list in word_lengths.items():
    print(f"Words of length {length}: {word_list}")
```

Question: Create a program to calculate the factorial of a number using recursion and list comprehension.

- Ask the user for a non-negative integer.
- Calculate the factorial of the number using a recursive function.
- Use list comprehension to generate a list of factorials from 0 to the given number.
- Display the list of factorials.

```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)

num = int(input("Enter a non-negative integer: "))

factorials = [factorial(i) for i in range(num + 1)]

print("Factorials:", factorials)
```

Question: Create a program to generate and display a Fibonacci sequence.

- Ask the user for the length of the sequence.
- Use a loop and the "If Structure" to generate the sequence.
- Display the sequence.

```python
def generate_fibonacci(length):
    fibonacci = [0, 1]
    while len(fibonacci) < length:
        next_num = fibonacci[-1] + fibonacci[-2]
        fibonacci.append(next_num)
    return fibonacci

length = int(input("Enter the length of the Fibonacci sequence: "))
fibonacci_sequence = generate_fibonacci(length)
print("Fibonacci sequence:", fibonacci_sequence)
```

Question: Create a program to simulate a simple bank account system.

Define a class BankAccount with methods for deposit, withdrawal, and balance check.
Use the "If Structure" to validate withdrawal amounts and ensure sufficient balance.
Ask the user for their choice and perform the corresponding action.

```python
class BankAccount:
    def __init__(self, balance=0):
        self.balance = balance
    
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
    
    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
    
    def check_balance(self):
        return self.balance

account = BankAccount()

while True:
    print("1. Deposit\n2. Withdraw\n3. Check Balance\n4. Quit")
    choice = int(input("Enter your choice: "))
    
    if choice == 1:
        amount = float(input("Enter the deposit amount: "))
        account.deposit(amount)
    elif choice == 2:
        amount = float(input("Enter the withdrawal amount: "))
        account.withdraw(amount)
    elif choice == 3:
        print("Current balance:", account.check_balance())
    elif choice == 4:
        print("Exiting...")
        break
```

Question: Create a program to analyze student grades.

- Define a dictionary to store student names as keys and their grades as values.
- Use the "If Structure" to categorize grades as 'Excellent', 'Good', 'Average', or 'Below Average'.
- Print a report showing each student's name, grade, and category.

```python
def categorize_grade(grade):
    if grade >= 90:
        return 'Excellent'
    elif grade >= 80:
        return 'Good'
    elif grade >= 70:
        return 'Average'
    else:
        return 'Below Average'

student_grades = {
    'Alice': 85,
    'Bob': 78,
    'Charlie': 92,
    'David': 63
}

print("Student Grade Report:")
for student, grade in student_grades.items():
    category = categorize_grade(grade)
    print(f"{student}: Grade {grade} ({category})")
```

Question: Create a program to analyze temperature data.

- Read temperature data from a CSV file.
- Calculate the average temperature and identify the hottest and coldest days.
- Use the "If Structure" to categorize temperatures as 'Hot', 'Moderate', or 'Cold'.
- Display the results.

```python
import csv

def categorize_temperature(temperature):
    if temperature > 30:
        return 'Hot'
    elif temperature > 15:
        return 'Moderate'
    else:
        return 'Cold'

def analyze_temperature(file_path):
    temperatures = []
    with open(file_path, 'r') as file:
        reader = csv.reader(file)
        next(reader)  # Skip header
        for row in reader:
            temperatures.append(float(row[1]))
    
    average_temperature = sum(temperatures) / len(temperatures)
    hottest_day = max(temperatures)
    coldest_day = min(temperatures)
    
    print("Temperature Analysis:")
    print(f"Average Temperature: {average_temperature:.2f}°C")
    print(f"Hottest Day: {hottest_day:.2f}°C")
    print(f"Coldest Day: {coldest_day:.2f}°C")
    
    for temperature in temperatures:
        category = categorize_temperature(temperature)
        print(f"{temperature:.2f}°C: {category}")

file_path = r"F:\My courses and projects\machine learning\datasets\Boston.csv"  # Replace with the path to your CSV file
analyze_temperature(file_path)
```

Question: Create a program to simulate a simple shopping cart.

- Define a class ShoppingCart with methods to add items, remove items, and calculate the total cost.
- Use the "If Structure" to apply discounts for certain items.
- Ask the user for their choice and perform the corresponding action.

```python
class ShoppingCart:
    def __init__(self):
        self.items = {}
    
    def add_item(self, item, price):
        if item in self.items:
            self.items[item] += price
        else:
            self.items[item] = price
    
    def remove_item(self, item):
        if item in self.items:
            del self.items[item]
    
    def calculate_total(self):
        total = sum(self.items.values())
        return total

cart = ShoppingCart()

while True:
    print("1. Add Item\n2. Remove Item\n3. Calculate Total\n4. Quit")
    choice = int(input("Enter your choice: "))
    
    if choice == 1:
        item = input("Enter the item name: ")
        price = float(input("Enter the item price: "))
        cart.add_item(item, price)
    elif choice == 2:
        item = input("Enter the item name: ")
        cart.remove_item(item)
    elif choice == 3:
        print("Total cost:", cart.calculate_total())
    elif choice == 4:
        print("Exiting...")
        break
```
