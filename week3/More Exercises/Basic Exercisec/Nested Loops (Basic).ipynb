Question: Generate a chessboard pattern (8x8 grid) using 'X' and 'O' characters.

```python
for i in range(8):
    for j in range(8):
        if (i + j) % 2 == 0:
            print('X', end=' ')
        else:
            print('O', end=' ')
    print()
```

Question: Create a program to solve the N-Queens puzzle using nested loops.

```python
def is_safe(board, row, col):
    for i in range(col):
        if board[row][i] == 1:
            return False
    for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False
    for i, j in zip(range(row, len(board), 1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False
    return True

def solve_n_queens(board, col):
    if col >= len(board):
        return True
    for i in range(len(board)):
        if is_safe(board, i, col):
            board[i][col] = 1
            if solve_n_queens(board, col + 1):
                return True
            board[i][col] = 0
    return False

def print_solution(board):
    for row in board:
        print(' '.join(map(str, row)))

n = 8
chessboard = [[0] * n for _ in range(n)]
if solve_n_queens(chessboard, 0):
    print_solution(chessboard)
else:
    print("No solution found.")
```

Question: Write a Python program to generate all possible valid IP addresses from a given string of digits.

```python
def is_valid(segment):
    return 0 <= int(segment) <= 255 and (len(segment) == 1 or segment[0] != '0')

def generate_ip_addresses(s):
    n = len(s)
    if n < 4 or n > 12:
        return []
    result = []
    for i in range(1, min(4, n - 2) + 1):
        for j in range(i + 1, min(i + 4, n - 1) + 1):
            for k in range(j + 1, min(j + 4, n) + 1):
                s1, s2, s3, s4 = s[:i], s[i:j], s[j:k], s[k:]
                if is_valid(s1) and is_valid(s2) and is_valid(s3) and is_valid(s4):
                    result.append(f"{s1}.{s2}.{s3}.{s4}")
    return result

digits = "25525511135"
ip_addresses = generate_ip_addresses(digits)
print(ip_addresses)
```

Question: Write a Python program to perform matrix multiplication using nested loops.

```python
matrix1 = [[1, 2], [3, 4]]
matrix2 = [[5, 6], [7, 8]]
result = [[0, 0], [0, 0]]
for i in range(len(matrix1)):
    for j in range(len(matrix2[0])):
        for k in range(len(matrix2)):
            result[i][j] += matrix1[i][k] * matrix2[k][j]
for row in result:
    print(row)
```

Question: Write a Python program to find and print all pairs of elements in an array that sum up to a specific target value.

```python
def find_pairs_with_sum(arr, target_sum):
    pairs = []
    for i in range(len(arr)):
        for j in range(i + 1, len(arr)):
            if arr[i] + arr[j] == target_sum:
                pairs.append((arr[i], arr[j]))
    return pairs

numbers = [2, 7, 4, 0, 9, 5, 1, 3]
target = 9
pairs = find_pairs_with_sum(numbers, target)
print(pairs)
```

Question: Create a program to generate a list of prime numbers within a specified range using nested loops.

```python
start = 10
end = 50
prime_numbers = []
for num in range(start, end + 1):
    if num > 1:
        is_prime = True
        for i in range(2, int(num ** 0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            prime_numbers.append(num)
print(prime_numbers)
```

Question: Write a Python program to find all possible combinations of three numbers from a given list that add up to a specific target sum.

```python
numbers = [1, 2, 3, 4, 5]
target_sum = 9
combinations = []
for i in range(len(numbers)):
    for j in range(i + 1, len(numbers)):
        for k in range(j + 1, len(numbers)):
            if numbers[i] + numbers[j] + numbers[k] == target_sum:
                combinations.append((numbers[i], numbers[j], numbers[k]))
print(combinations)
```

Question: Create a program to find and print all unique triplets from an array that add up to a specific target sum.

```python
arr = [1, 2, 3, 4, 5, 6]
target_sum = 9
triplets = set()
for i in range(len(arr) - 2):
    for j in range(i + 1, len(arr) - 1):
        for k in range(j + 1, len(arr)):
            if arr[i] + arr[j] + arr[k] == target_sum:
                triplets.add(tuple(sorted([arr[i], arr[j], arr[k]])))
print(list(triplets))
```

Question: Write a Python program to find and print all anagrams in a given list of words.

```python
words = ["listen", "silent", "hello", "world", "python", "typhon"]
anagrams = {}
for word in words:
    sorted_word = ''.join(sorted(word))
    if sorted_word in anagrams:
        anagrams[sorted_word].append(word)
    else:
        anagrams[sorted_word] = [word]
for key in anagrams:
    if len(anagrams[key]) > 1:
        print(anagrams[key])
```
